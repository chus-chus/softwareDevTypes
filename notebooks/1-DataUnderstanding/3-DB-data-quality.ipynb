### A brief visual exploration of missing values on the different tables.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_PATH = '../../data/raw/'

# No need to extract relationships in this phase, so plain csv reading is good enough.
# Beware: have at least 8GB of memory

projectsDf = pd.read_csv(DATA_PATH + 'PROJECTS.csv')
jiraIssuesDf = pd.read_csv(DATA_PATH + 'JIRA_ISSUES.csv')
sonarMeasuresDf = pd.read_csv(DATA_PATH + 'SONAR_MEASURES.csv')
sonarIssuesDf = pd.read_csv(DATA_PATH + 'SONAR_ISSUES.csv')
sonarRulesDf = pd.read_csv(DATA_PATH + 'SONAR_ISSUES.csv')
gitCommitsDf = pd.read_csv(DATA_PATH + 'GIT_COMMITS.csv')
szzFaultInducingCommitsDf = pd.read_csv(DATA_PATH + 'SZZ_FAULT_INDUCING_COMMITS.csv')
gitCommitsChangesDf = pd.read_csv(DATA_PATH + 'GIT_COMMITS_CHANGES.csv')
refactoringMinerDf = pd.read_csv(DATA_PATH + 'REFACTORING_MINER.csv')
```

```python
def missing_stats(df, title):
    """ Plots the pct. of null values of all features in a DataFrame. """
    numberOfNullsSeries = df.isnull().sum()
    pctNulls = numberOfNullsSeries / len(df)
    pctNulls.plot.bar(title='Percentage of null entries in Df ' + title)
    plt.show(block=True)

missing_stats(jiraIssuesDf, 'jiraIssues')
```

```python
missing_stats(gitCommitsChangesDf, 'git Commit Changes')
```
