# **TIME_IN_EACH_PROJECT**

This notebook the creation of the table `TIME_IN_EACH_PROJECT`, that contains the time that each committer has been worked in each project.

First, we import the libraries we need and, then, we read the corresponding csv.

```python
import pandas as pd
```

```python
gitCommits = pd.read_csv("../../../data/interim/DataPreparation/CleanData/GIT_COMMITS_clean.csv")
print(gitCommits.shape)
gitCommits.head()
```

First, we get the time of the first and last commit from each committer in each project.

```python
time_in_project = gitCommits.groupby(['projectID', 'committer'])['committerDate'].agg(['min', 'max']).reset_index()
time_in_project
```

Then, we calculate the difference between these 2 values to know how long that developer has been in a project.

```python
time = (pd.to_datetime(time_in_project['max'])-pd.to_datetime(time_in_project['min']))
time_in_project['time'] = time.dt.total_seconds()
```

```python
time_in_project
```

```python
time_in_project = time_in_project[['projectID', 'committer', 'time']]
time_in_project
```

We save this table in a new csv.

```python
time_in_project.to_csv('../../../data/interim/DataPreparation/ConstructData/TIME_IN_PROJECT.csv', header=True)
```
